```python
a = 1
b = 2
if a < b:
    print("a<b")
    print("a less than b")
elif a== b:
    print("a equal b")
else:
    print("a more than b")
```

**10. Function**

```python
def power(a, b):
    c = a**b
    return c

power(2,3)
```

```python
def summation(L):
    z = 0
    for item in L:
        z = z + item
    return z

summation([1, 2, 3])
```

```python
def multi(start, end):
    for i in range(start, end+1):
         for j in range(12):
            print(str(i) + "x" + str(j+1) + "=" + str(i*(j+1)))
         print("\n") 
    
multi(3, 5)
```
